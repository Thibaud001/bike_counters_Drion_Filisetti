```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
external_data = pd.read_csv('external_data.csv')
```

```python
external_data_cleaned = external_data.dropna(axis=1, how='all')
```

```python
columns_of_interest = ["date", "etat_sol", "dd", "ff", "t", "u", "vv", "n", "ht_neige", "rr3"]

external_data_sorted = external_data_cleaned[columns_of_interest]
external_data_sorted.loc[: ,"t"] = external_data_sorted["t"] - 273.15
external_data_sorted.head()
```

```python
# Convertir la colonne 'date' en datetime en utilisant .loc
external_data_sorted["date"] = pd.to_datetime(external_data_cleaned['date'])
```

```python
def _encode_dates(X):
    X = X.copy()  # modify a copy of X
    # Encoder les informations de date à partir de la colonne 'date'
    X["year"] = X["date"].dt.year
    X["month"] = X["date"].dt.month
    X["day"] = X["date"].dt.day
    X["weekday"] = X["date"].dt.weekday
    X["hour"] = X["date"].dt.hour

    # Supprimer la colonne 'date' originale
    return X.drop(columns=["date"])

external_data_sorted = _encode_dates(external_data_sorted)
external_data_sorted.head()
```
